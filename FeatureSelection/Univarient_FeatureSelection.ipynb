Univariant Selection: This can be used to select features based on having the strongest relation ship with features

Using Chi-squared we can select K best features from a given data set

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
m_data = pd.read_csv('C:\\Users\\Rahul\\Desktop\\ML\\KrishNaik\\Feature-Selection-techniques\\m_train.csv')
m_data.head()
```

Differentiation Dependent and Independent Variables

```python
m_data.shape
```

```python
x = m_data.iloc[:,0:20]
y = m_data.iloc[:, -1]
```

```python
best_features = SelectKBest(score_func=chi2, k =10)
```

```python
fit = best_features.fit(x,y)
```

```python
fit.scores_
```

```python
df_scores = pd.DataFrame(fit.scores_)
```

```python
df_scores
```

```python
df_columns = pd.DataFrame(x.columns)
```

```python
df_columns
```

```python
feature_scores = pd.concat([df_columns, df_scores], axis= 1)
```

```python
feature_scores
```

```python
feature_scores.columns = ['specs', 'feature_score']
```

```python
feature_scores
```

```python
feature_scores.nlargest(11, 'feature_score')
```
